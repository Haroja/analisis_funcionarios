# Importar librerias

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

# importar archivo csv 

```python
df_funcionarios = pd.read_csv("remuneración_autoridades_superiores_junio_2024.csv", encoding='ISO-8859-1', delimiter=';')
```

# Exploración de datos

```python
df_funcionarios.head()
```

# Limpieza de datos

```python

# Reemplazar comas por puntos solo en columnas que contienen números
for col in df_funcionarios.columns:
    if df_funcionarios[col].dtype == "object":
        
        try:
            df_funcionarios[col] = df_funcionarios[col].str.replace(',', '.').astype(float)
        except ValueError:
            
            print(f"Columna '{col}' no convertida a float. Contiene valores no numéricos.")


df_funcionarios.head()
```

```python
df_funcionarios.drop_duplicates()
```

```python
df_funcionarios.dropna()
```

```python
# Deshabilitar la notación científica en pandas
pd.set_option('display.float_format', '{:.2f}'.format)
print(df_funcionarios.head())
```

# Información estadistica

```python
desc_stats = df_funcionarios.describe()

desc_stats_T = desc_stats.T

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
fig.suptitle("Estadísticas Descriptivas de df_funcionarios", fontsize=16)

desc_stats_T["mean"].plot(kind="bar", ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title("Media")
axes[0, 0].set_ylabel("Valor")

desc_stats_T["std"].plot(kind="bar", ax=axes[0, 1], color="salmon")
axes[0, 1].set_title("Desviación Estándar")
axes[0, 1].set_ylabel("Valor")


percentiles = desc_stats_T[["25%", "50%", "75%"]]
percentiles.plot(kind="bar", ax=axes[1, 0], colormap="coolwarm")
axes[1, 0].set_title("Percentiles")
axes[1, 0].set_ylabel("Valor")

min_max = desc_stats_T[["min", "max"]]
min_max.plot(kind="bar", ax=axes[1, 1], color=["lightgrey", "darkgrey"])
axes[1, 1].set_title("Mínimo y Máximo")
axes[1, 1].set_ylabel("Valor")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

# Nueva columna con el porcentaje de descuentos sobre el básico

```python
df_funcionarios["PORCENTAJE_DESCUENTO_BASICO"] = df_funcionarios["DESCUENTOS DE LEY (19,3%)"] / df_funcionarios[ 'BASICO A PARTIR DEL 01/06/2024    '] * 100

df_funcionarios[["DENOMINACION", "PORCENTAJE_DESCUENTO_BASICO"]].head()
```

# Comparación entre los cargos y haber neto

```python
# Agrupar por cargos y calcular el promedio del haber neto
promedio_haber_neto_por_cargo = df_funcionarios.groupby("CARGOS")["HABER NETO"].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=promedio_haber_neto_por_cargo.index, y=promedio_haber_neto_por_cargo.values, hue=promedio_haber_neto_por_cargo.index, palette="coolwarm", legend=False)
plt.title("Promedio del Haber Neto por Cargo")
plt.xlabel("Cargos")
plt.ylabel("Promedio del Haber Neto")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# Haber neto por cargos

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="CARGOS", y="HABER NETO", data=df_funcionarios)
plt.title("Haber Neto por Cargo")
plt.xticks(rotation=45)
plt.show()
```

# Histograma haber neto

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_funcionarios["HABER NETO"], bins=30, kde=True)
plt.title("Distribución del Haber Neto")
plt.xlabel("Haber Neto")
plt.ylabel("Frecuencia")
plt.show()
```

# Haber neto mayores a 1800000.00

```python

df_filtrado = df_funcionarios[df_funcionarios['HABER NETO'] > 1800000]
plt.figure(figsize=(10, 6))
plt.bar(df_filtrado['DENOMINACION'], df_filtrado['HABER NETO'], color="skyblue")
plt.title("Denominaciones con Haber Neto > 1.800.000", fontsize=16)
plt.xlabel("Denominación", fontsize=12)
plt.ylabel("Haber Neto", fontsize=12)
plt.xticks(rotation=90)  
plt.tight_layout()

plt.show()
```

# ¿Cuál es la relación entre el haber neto y el total?

```python
plt.figure(figsize=(10, 6))
sns.regplot(x="TOTAL", y="HABER NETO", data=df_funcionarios, color='#FF00FF', line_kws={"color": "blue"})
plt.title("Relación entre Total y Haber Neto")
plt.xlabel("Total")
plt.ylabel("Haber Neto")
plt.show()

correlacion_total_haber_neto = df_funcionarios["TOTAL"].corr(df_funcionarios["HABER NETO"])
print(f"La correlación entre el total y el haber neto es: {correlacion_total_haber_neto}")
```

# ¿Cuál es el cargo con el mayor promedio de haber neto?

```python
promedio_haber_neto_por_cargo = df_funcionarios.groupby("CARGOS")["HABER NETO"].mean()

cargo_mayor_promedio_haber_neto = promedio_haber_neto_por_cargo.idxmax()
promedio_maximo = promedio_haber_neto_por_cargo.max()

df_promedios = pd.DataFrame({
    "CARGOS": promedio_haber_neto_por_cargo.index,
    "PROMEDIO_HABER_NETO": promedio_haber_neto_por_cargo.values
})

plt.figure(figsize=(12, 8))
sns.barplot(x="CARGOS", y="PROMEDIO_HABER_NETO", data=df_promedios, palette="viridis", hue="CARGOS")


plt.axhline(promedio_maximo, color="red", linestyle="--", linewidth=2, label=f'{cargo_mayor_promedio_haber_neto} (Promedio Máximo)')
plt.legend()

for i, row in df_promedios.iterrows():
    plt.text(i, row["PROMEDIO_HABER_NETO"] + 0.5, f'{row["PROMEDIO_HABER_NETO"]:.2f}', ha="center", va="bottom")

plt.xlabel("Cargos", fontsize=14)
plt.ylabel("Promedio de Haber Neto", fontsize=14)
plt.title("Promedio de Haber Neto por Cargo", fontsize=16)
plt.xticks(rotation=90)  

plt.tight_layout()
plt.show()

print(f"El cargo con mayor promedio de haber neto es: {cargo_mayor_promedio_haber_neto} con un promedio de {promedio_maximo:.2f}")
```

# ¿Qué porcentaje del total se descuenta en promedio por ley?

```python
porcentaje_promedio_descuento = (df_funcionarios["DESCUENTOS DE LEY (19,3%)"] / df_funcionarios["TOTAL"]).mean() * 100
valores = [porcentaje_promedio_descuento, 100 - porcentaje_promedio_descuento]
etiquetas = ["Descuento por Ley (19.3%)", "Resto"]
plt.figure(figsize=(6, 6))
plt.pie(valores, labels=etiquetas, autopct='%1.1f%%', colors=["#FF00FF", "#ADD8E6"], startangle=90)
plt.title("Porcentaje Promedio de Descuento Aplicado por Ley")
plt.show()
print(f"El porcentaje promedio de descuento aplicado por ley es: {porcentaje_promedio_descuento:.2f}%")
```

# ¿Qué porcentaje del basico a partir del 1/06/2024 corresponde al descuento de ley (19.3%) para cada denominación?

```python
df_funcionarios_seleccionado = df_funcionarios[["DENOMINACION", "PORCENTAJE_DESCUENTO_BASICO"]]

# Eliminar las filas con valores nulos en las columnas seleccionadas
df_funcionarios_limpio = df_funcionarios_seleccionado.dropna()

# Mostrar los primeros resultados del DataFrame limpio
print(df_funcionarios_limpio.head())
```

# ¿Cuál es el denominación que tiene el mayor descuento en porcentaje sobre el total?

```python
denominacion_mayor_descuento = df_funcionarios[["DENOMINACION", "PORCENTAJE_DESCUENTO_BASICO"]].sort_values(by="PORCENTAJE_DESCUENTO_BASICO", ascending=False).iloc[0]

plt.figure(figsize=(8, 6))
plt.bar(denominacion_mayor_descuento["DENOMINACION"], denominacion_mayor_descuento["PORCENTAJE_DESCUENTO_BASICO"], color="teal")

plt.xlabel("Denominación", fontsize=14)
plt.ylabel("Porcentaje de Descuento Básico (%)", fontsize=14)
plt.title("Denominación con Mayor Porcentaje de Descuento Básico", fontsize=16)
plt.xticks(rotation=45, ha="right")  
plt.tight_layout()
plt.show()

print(f"La denominación con mayor descuento es: {denominacion_mayor_descuento}")
```
